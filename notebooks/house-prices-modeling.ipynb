```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows',100)
```

# Data Setup

**Load the data**

```python
data = pd.read_csv("../data/train.csv")
```

```python
data.head(20)
```

```python
data.shape
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].astype('category')
```

```python
X = data.drop(columns=['SalePrice'])
y = data['SalePrice']
```

**Split data into train and test sets**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Feature selection

```python
num_cols = ['LotArea','LotFrontage','MasVnrArea','TotalBsmtSF','GrLivArea','GarageYrBlt','GarageArea','BsmtFinSF1','BsmtFinSF2','1stFlrSF','2ndFlrSF',
            'WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea']
cat_cols = ['LotShape','LotConfig','BldgType','HouseStyle','MasVnrType','BsmtQual','GarageType','GarageFinish','RoofStyle','Foundation','SaleCondition','Utilities','LandSlope']
```

# Feature processing

**Checking missing values**

```python
print("Missing Values for X_Train in Numerical Columns")
print(X_train[num_cols].isna().sum())
```

```python
print("Missing Values for X_Train in Categorical Columns")
print(X_train[cat_cols].isna().sum())
```

```python
print("Missing Values for X_test in Numerical Columns")
print(X_test[num_cols].isna().sum())
```

```python
print("Missing Values for X_test in Categorical Columns")
print(X_test[cat_cols].isna().sum())
```

**Filling Missing Values in Train and Test Sets**  
- Numerical Columns: Fill with **Mean**  
- Categorical Columns: Fill with **Mode (Most Frequent)**  

```python
for dataset in [X_train, X_test]:
    dataset.fillna({col: dataset[col].mean() for col in num_cols}, inplace=True)
    dataset.fillna({col: dataset[col].mode()[0] for col in cat_cols}, inplace=True)
```

**After filling missing values**

```python
print("After fillng Values for X_Train in Numerical Columns")
print(X_train[num_cols].isna().sum())
```

```python
print("After fillng Values for X_Train in Categorical Columns")
print(X_train[cat_cols].isna().sum())
```

```python
print("After fillng Values for X_test in Numerical Columns")
print(X_test[num_cols].isna().sum())
```

```python
print("After fillng Values for X_test in Categorical Columns")
print(X_test[cat_cols].isna().sum())
```

```python
print("Missing Values for y_train")
print(y_train.isna().sum())
```

```python
print("Missing Values for y_test")
print(y_train.isna().sum())
```

**Scaling Numerical Features - StandardScaler**

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
std.fit(X_train[num_cols])

X_train_num = std.transform(X_train[num_cols])
X_test_num = std.transform(X_test[num_cols])
```

```python
X_train_scaled = pd.DataFrame(X_train_num, columns=num_cols)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(X_test_num, columns=num_cols)
X_test_scaled
```

**Encoding Categorical Features - OneHotEncoder**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X_train[cat_cols])

X_train_cat = ohe.transform(X_train[cat_cols]).toarray()
X_test_cat = ohe.transform(X_test[cat_cols]).toarray()
```

```python
encoded_cols = ohe.get_feature_names_out(cat_cols)
```

```python
X_train_encoded = pd.DataFrame(X_train_cat, columns=encoded_cols)
X_train_encoded
```

```python
X_test_encoded = pd.DataFrame(X_test_cat, columns=encoded_cols)
X_test_encoded
```

**Merging the data ready for Model Training**

```python
X_train_final = pd.concat([X_train_scaled, X_train_encoded], axis=1)
X_train_final
```

```python
X_test_final = pd.concat([X_test_scaled, X_test_encoded], axis=1)
X_test_final
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train_final,y_train)
```

```python
y_train_preds = model.predict(X_train_final)
y_test_preds = model.predict(X_test_final)
```

# Model evaluation

**Performance Analysis with R2-Score**

```python
from sklearn.metrics import r2_score

train_score = r2_score(y_train,y_train_preds)
test_score = r2_score(y_test,y_test_preds)
print("Train Score: ",f'{train_score:.2f}')
print("Test Score: ",f'{test_score:.2f}')
```

**Computation Metric with RMSE (Root-Mean-Squared-Error)**

```python
from sklearn.metrics import mean_squared_error

def compute_rmse(test,pred):
    rmse = np.sqrt(mean_squared_error(test,pred))
    return f'{rmse:2f}'
```

```python
print("RMSE for Train Data")
print(compute_rmse(y_train,y_train_preds))
print('------------------------------------')
print("RMSE for Test Data")
print(compute_rmse(y_test,y_test_preds))
```
